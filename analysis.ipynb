#### Player performance model

```python
c_df.columns
```

```python
c_df[['FotMob_rating', 'xG', 'Expected_assists_xA', 'Goals', 'Assists', 'Accurate_passes_value', 'Defensive_actions']]
```

```python
# select features
features = [
    'FotMob_rating',
    'xG',
    'Expected_assists_xA',
    'Goals',
    'Assists',
    'Accurate_passes_value',
    'Defensive_actions'
]
```

```python
# clean numeric values

model_df = c_df[features].copy()

# Convert everything to numeric
for col in model_df.columns:
    model_df[col] = pd.to_numeric(model_df[col], errors='coerce')

model_df = model_df.fillna(0)
```

```python
model_df
```

```python
# normalize
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(model_df)

scaled_df = pd.DataFrame(scaled, columns=model_df.columns)
```

```python
# create weighted scored
weights = {
    'FotMob_rating': 0.30,
    'Goals': 0.20,
    'Assists': 0.15,
    'xG': 0.15,
    'Expected_assists_xA': 0.10,
    'Accurate_passes_value': 0.05,
    'Defensive_actions': 0.05
}
```

```python
scaled_df['performance_score'] = sum(
    scaled_df[col] * weight for col, weight in weights.items()
)
```

```python
# attach to original data
final_df = pd.concat([c_df, scaled_df['performance_score']], axis=1)

final_df.sort_values('performance_score', ascending=False).head()
```

```python
# dashboard matplotlib
import matplotlib.pyplot as plt

top_players = final_df.sort_values('performance_score', ascending=False).head(10)

plt.figure()
plt.barh(top_players.index.astype(str), top_players['performance_score'])
plt.xlabel("Performance Score")
plt.title("Top 10 Players")
plt.gca().invert_yaxis()
plt.show()
```

```python
# xG vs Goals (efficiency)
plt.figure()
plt.scatter(
    final_df['xG'],
    final_df['Goals']
)

plt.xlabel("xG")
plt.ylabel("Goals")
plt.title("Finishing Efficiency")

plt.show()
```

```python
# FotMob Rating vs Performance Score
plt.figure()
plt.scatter(
    final_df['FotMob_rating'],
    final_df['performance_score']
)

plt.xlabel("Rating")
plt.ylabel("Performance Score")
plt.title("Model vs FotMob Rating")

plt.show()
```
